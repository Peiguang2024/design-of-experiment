### Exercise 17.2

```python
r = []
distance = []
for angle in np.linspace(0,np.pi/2,6, endpoint=False):
    r.append(1)
    if angle >= np.pi/4:
        distance.append(1 * np.cos(np.pi/4) / np.cos(np.pi/2-angle))
    else:
        distance.append(1 * np.cos(np.pi/4) / np.cos(angle))

np.tile(r,4), np.tile(distance,4)
```

```python
plt.scatter(np.linspace(0, np.pi * 2, 24, endpoint=False), (np.tile(r,4) - np.tile(distance,4)))
```
