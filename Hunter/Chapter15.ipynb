### Exercise15.3

```python
df_t1501 = pd.DataFrame({
    'x1': [-1,1,-1,1,0,0,0],
    'x2': [-1,-1,1,1,0,0,0],
    'yield': [54.3,60.3,64.6,68.0,60.3,64.3,62.3]
})


df_t1501
```

```python
df_e1503 = df_t1501.copy()
df_e1503['yield'] += 10
df_e1503
```

```python
df_e1503['yield'].mean() # b0
```

```python
df_e1503[['x1','yield']].prod(axis=1).sum()/2/2 # b1
```

```python
df_e1503[['x2','yield']].prod(axis=1).sum()/2/2 # b1
```

```python
arr_e1503_x = np.concat((np.array([1]*7).reshape(7,1), df_e1503[['x1','x2']].to_numpy()), axis=1)
arr_e1503_y = df_e1503[['yield']].to_numpy()

arr_e1503_x, arr_e1503_y
```

```python
np.linalg.inv(arr_e1503_x.T.dot(arr_e1503_x)).dot(arr_e1503_x.T.dot(arr_e1503_y))
```

```python
arr_e1503_x.dot(np.linalg.inv(arr_e1503_x.T.dot(arr_e1503_x)).dot(arr_e1503_x.T.dot(arr_e1503_y))) # y_hat
```

```python
np.sqrt(np.linalg.inv(arr_e1503_x.T.dot(arr_e1503_x))*2.25) # standard errors
```

### Problem 15.5

```python
df_e1506 = pd.DataFrame({
    'x1': [-np.sqrt(2), np.sqrt(2), 0, 0, 0, 0],
    'x2': [0, 0, -np.sqrt(2), np.sqrt(2), 0, 0],
    'yield': [83.3,81.2,81.2,79.5,87.0,86.0],
    })

df_e1506 = pd.concat((df_e1505.iloc[:,:3], df_e1506)).reset_index(drop=True)
df_e1506
```

```python
df_e1506['x11'] = df_e1506['x1']**2
df_e1506['x12'] = df_e1506[['x1', 'x2']].prod(axis=1)
df_e1506['x22'] = df_e1506['x2']**2
df_e1506['x0'] = 1
```

```python
X_e1506 = df_e1506[['x0', 'x1', 'x2', 'x11', 'x12', 'x22']].to_numpy()
y_e1506 = df_e1506[['yield']].to_numpy()
```

```python
np.linalg.inv(X_e1506.T.dot(X_e1506)).dot(X_e1506.T.dot(y_e1506)) #b0 b1 b2 b11 b12 b22
```

```python
(np.linalg.inv(X_e1506.T.dot(X_e1506))*2.25).round(3)#variance-covariance matrix
```
